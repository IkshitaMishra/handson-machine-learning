# K-Means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Dataset

```python
dataset = pd.read_csv(r'C:\Users\Ikshita\Downloads\K_Means\Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
```

# Finding 'k' using Elbow Method

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('the Elbow Method')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```

# Building Model to Training Data

```python
kmeans = KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
```

# Prediction

```python
y_kmeans = kmeans.fit_predict(x)
```

# Visualization



```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='yellow',label='Cluster 4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='black',label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='pink',label='Centroid')
plt.title('Cluster of clients')
plt.xlabel('Annual Income')
plt.ylabel('Spending score')
plt.show()
```
