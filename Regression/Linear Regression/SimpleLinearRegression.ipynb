# Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Dataset

```python
dataset = pd.read_csv(r"C:\Users\Ikshita\Downloads\Simple_Linear_Regression\Simple_Linear_Regression\StockPrice.csv")
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
```

# Splitting Dataset

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state = 0)
```

# Building and Fitting Model to Training Data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

# Prediction

```python
y_pred = regressor.predict(x_test)
```

# Visualization


## Training Set

```python
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, regressor.predict(x_train))
plt.title('Stock Price vs Years (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Stock Price ')
plt.show()
```

# Test Set

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, regressor.predict(x_train))
plt.title('Stock Price vs Years (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Stock Price ')
plt.show()
```
